```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,50]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label="max",marker="*")
plt.plot(days,min_t,label="min",marker="*")
plt.plot(days,avg_t,label="avg",marker="*")
plt.xlabel("days")
plt.ylabel("temperature")

plt.legend(loc="best")
plt.grid()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profits=[60,46,65,20]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.bar(ypos-0.2,revenue,label="revenue",width=0.4)
plt.bar(ypos+0.2,profits,label="profits",width=0.4)

plt.legend()
plt.savefig("barchart.png")
```
