```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\G Sai Madhav\Desktop\robotics\ML\ONE_HOT_ENCODING\one_hot_encoding.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
```

```python
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
final = merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
reg.fit(final[['area','monroe township','robinsville']],final.price)
```

```python
reg.predict([[2800,0,1]])
```

```python
reg.predict([[3400,0,0]])
```

```python
import pickle

with open('one_hot_enc','wb') as f:
    pickle.dump(reg,f)
```
